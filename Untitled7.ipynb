# Exercício 1

```python
# Dados do problema
acertos = 8
total = 10
prob_acerto = acertos / total

# 1. Probabilidade de errar uma previsão
prob_erro = 1 - prob_acerto
print(f"a) Probabilidade de errar uma previsão: {prob_erro:.2f} ou {prob_erro*100:.0f}%")

# 2. Probabilidade de acertar 5 previsões consecutivas
prob_5_acertos = prob_acerto ** 5
print(f"b) Probabilidade de acertar 5 previsões consecutivas: {prob_5_acertos:.4f} ou {prob_5_acertos*100:.2f}%")

# 3. Interpretação
print("\nc) Interpretação:")
print(f"- O modelo tem acurácia de {prob_acerto*100:.0f}%")
print(f"- Em 5 previsões independentes, espera-se acertar {prob_acerto * 5:.1f} em média")
print(f"- A chance de acertar todas as 5 é de apenas {prob_5_acertos*100:.1f}%")
print("- Isso demonstra que mesmo modelos bons podem errar previsões consecutivas")
print("- Em aplicações críticas, é importante considerar margens de erro")
```

# Exercício 2

```python
import random
import numpy as np
import matplotlib.pyplot as plt

# Simulação com moeda justa (p=0.5)
print("=== MOEDA JUSTA (p=0.5) ===")
tamanhos = [10, 100, 1000, 10000]
proporcoes_justa = []

for n in tamanhos:
    # Usando random.choice para moeda justa
    resultados = [random.choice(["Cara", "Coroa"]) for _ in range(n)]
    p_cara = resultados.count("Cara") / n
    proporcoes_justa.append(p_cara)
    print(f"n = {n:5d}: Proporção de Caras = {p_cara:.4f}")

# Simulação com moeda viciada (p=0.2)
print("\n=== MOEDA VICIADA (p=0.2) ===")
proporcoes_viciada = []

for n in tamanhos:
    # Usando numpy.random.choice com probabilidades diferentes
    resultados = np.random.choice(["Cara", "Coroa"], size=n, p=[0.2, 0.8])
    p_cara = np.sum(resultados == "Cara") / n
    proporcoes_viciada.append(p_cara)
    print(f"n = {n:5d}: Proporção de Caras = {p_cara:.4f}")

# Gráficos comparativos
plt.figure(figsize=(14, 5))

# Gráfico 1: Moeda justa
plt.subplot(1, 2, 1)
plt.plot(tamanhos, proporcoes_justa, 'bo-', markersize=8, linewidth=2, label='Proporção observada')
plt.axhline(0.5, color='red', linestyle='--', linewidth=2, label='Probabilidade teórica = 0.5')
plt.xscale('log')
plt.xlabel('Número de lançamentos (escala logarítmica)')
plt.ylabel('Proporção de Caras')
plt.title('Lei dos Grandes Números - Moeda Justa\n(random.choice)')
plt.legend()
plt.grid(True, alpha=0.3)

# Gráfico 2: Moeda viciada
plt.subplot(1, 2, 2)
plt.plot(tamanhos, proporcoes_viciada, 'go-', markersize=8, linewidth=2, label='Proporção observada')
plt.axhline(0.2, color='red', linestyle='--', linewidth=2, label='Probabilidade teórica = 0.2')
plt.xscale('log')
plt.xlabel('Número de lançamentos (escala logarítmica)')
plt.ylabel('Proporção de Caras')
plt.title('Lei dos Grandes Números - Moeda Viciada\n(np.random.choice com p=[0.2, 0.8])')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Explicação da Lei dos Grandes Números
print("\n=== EXPLICAÇÃO DA LEI DOS GRANDES NÚMEROS ===")
print("A Lei dos Grandes Números estabelece que:")
print("1. Com poucos experimentos (n=10): a proporção observada varia bastante")
print("2. Com mais experimentos (n=100, 1000): a proporção se estabiliza")
print("3. Com muitos experimentos (n=10000): a proporção se aproxima da probabilidade teórica")
print("\nIsso é fundamental em Machine Learning porque:")
print("- Conjuntos de dados maiores fornecem estimativas mais precisas")
print("- A performance medida em dados de teste é mais confiável com amostras grandes")
print("- Parâmetros aprendidos convergem para valores mais próximos da realidade")
```

# Exercício 3

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Parâmetros da distribuição
mu = 2.0      # média
sigma = 0.3   # desvio padrão
n_amostras = 1000

# 1. Gerando 1.000 amostras usando numpy.random.normal
np.random.seed(42)
tempos = np.random.normal(mu, sigma, n_amostras)

print(f"Parâmetros: μ = {mu}, σ = {sigma}")
print(f"Número de amostras: {n_amostras}")

# 2. Plotando histograma e curva teórica
plt.figure(figsize=(12, 5))

# Histograma das amostras
plt.hist(tempos, bins=30, density=True, alpha=0.7, color='lightblue',
         label='Amostras simuladas (histograma)')

# Curva teórica da Normal usando scipy.stats.norm
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y_teorica = norm.pdf(x, mu, sigma)  # PDF teórica
plt.plot(x, y_teorica, 'r-', linewidth=2, label='Curva teórica N(2.0, 0.3)')

# 3. Calculando probabilidade de tempo > 2.5 segundos usando scipy.stats.norm
prob_maior_25 = 1 - norm.cdf(2.5, mu, sigma)  # 1 - P(X ≤ 2.5)

print(f"\nProbabilidade do tempo ser maior que 2.5 segundos:")
print(f"Usando scipy.stats.norm.cdf: {prob_maior_25:.4f} ou {prob_maior_25*100:.2f}%")

# Destacando a região > 2.5 segundos no gráfico
x_maior = np.linspace(2.5, mu + 4*sigma, 100)
y_maior = norm.pdf(x_maior, mu, sigma)
plt.fill_between(x_maior, y_maior, alpha=0.3, color='red',
                label=f'P(X > 2.5) = {prob_maior_25:.3f}')

plt.axvline(2.5, color='red', linestyle='--', alpha=0.7, label='Limite de 2.5s')
plt.axvline(mu, color='green', linestyle='--', alpha=0.7, label=f'Média μ = {mu}')

plt.xlabel('Tempo de resposta (segundos)')
plt.ylabel('Densidade de probabilidade')
plt.title('Distribuição Normal do Tempo de Resposta do Sensor\n' +
          f'(numpy.random.normal + scipy.stats.norm)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# 4. Análise adicional das amostras
print(f"\nAnálise das amostras geradas com numpy.random.normal:")
print(f"Média das amostras: {np.mean(tempos):.3f} segundos")
print(f"Desvio padrão das amostras: {np.std(tempos):.3f} segundos")
print(f"Valor mínimo: {np.min(tempos):.3f} segundos")
print(f"Valor máximo: {np.max(tempos):.3f} segundos")

# Verificação empírica da probabilidade
tempos_maiores_25 = tempos[tempos > 2.5]
prob_empirica = len(tempos_maiores_25) / len(tempos)
print(f"\nVerificação empírica:")
print(f"Amostras com tempo > 2.5s: {len(tempos_maiores_25)} de {n_amostras}")
print(f"Probabilidade empírica: {prob_empirica:.4f} ou {prob_empirica*100:.2f}%")

# 5. Explicação da aplicação em IA
print(f"\nAplicação em IA - Modelagem de Ruído em Sensores:")
print("As distribuições probabilísticas são úteis para:")
print("a) Modelar variações naturais no tempo de resposta de sensores")
print("b) Calcular probabilidades de falha ou desempenho inadequado")
print("c) Definir limites de tolerância para sistemas em tempo real")
print("d) Simular cenários realistas para teste de algoritmos")
print("e) Implementar filtros probabilísticos (ex: Filtro de Kalman)")
print(f"\nNeste caso, sabemos que há {prob_maior_25*100:.1f}% de chance do sensor")
print(f"ter um tempo de resposta acima do limite de 2.5 segundos.")
```
